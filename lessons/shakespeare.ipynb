# quantifying shakespeare

downloading the plain text corpus of shakespeare's work 
 from < bit.ly/shakespeare-txt > 

## Plan of action

1. read every line of text into a list of strings
2. count the frequencies of each letter
3. visualize the frequencies

```python
# read lines in from file
from io import TextIOWrapper

def read_lines(filename:str) -> list[str]:
    """Reading every line fromfilename into a list of str."""
    lines: list[str] = []
    # TODO
    file_handle: TextIOWrapper = open(filename, "r")
    for line in file_handle:
        line = line.lower()
        line = line.strip()
        lines.append(line)
    return lines

shakespeare_lines: list[str] = read_lines("./shakespeare.txt")
print(len(shakespeare_lines))
print(shakespeare_lines[0])
```

```python
#define function named tally
#2 parameters: 
def tally(input: dict[str, int], letter: str) -> None:
    
    if letter in input:
        input[letter] += 1
    else: 
        input[letter] = 1
    return None

def count_letters(lines: list[str]) -> dict[str,int]:
    counts: dict[str, int] = {}
    for line in lines:
        for letter in line:
            if letter >= "a" and letter <= "z":
                tally(counts, letter)

    return counts

shakespeare_letters: dict[str, int] = count_letters(shakespeare_lines)
print(shakespeare_letters)
```

```python
import matplotlib.pyplot as plt
plt.title("FRequency of letters")
plt.xlabel("Letters")
plt.ylabel("Frequency")
labels = list(shakespeare_letters.keys())
values = list(shakespeare_letters.values())
plt.bar(labels, values)
plt.show()
```
